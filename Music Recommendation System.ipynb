```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier

music_data = pd.read_csv('music.csv')      #Importing Data
X = music_data.drop(columns=['genre'])     #Input Set 
y = music_data['genre']

model = DecisionTreeClassifier()
model.fit(X,y)
predictions = model.predict([[21,1],[22,0]])
predictions
```
